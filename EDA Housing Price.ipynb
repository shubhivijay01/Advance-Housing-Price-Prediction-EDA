```python
## Data Analysis Phase
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

pd.pandas.set_option('display.max_columns',None) 
```

```python
dataset=pd.read_csv('train.csv')
print(dataset.shape)
```

```python
dataset.head()
```

## Missing Values

```python
dataset.isnull().values.any()
feature_with_na=[features for features in dataset.columns if dataset[features].isnull().sum()>1]

for feature in feature_with_na:
    print(feature, np.round(dataset[feature].isnull().mean(), 4), '% missing values')  
     #% of missing values
```

### Since they are many missing values, we need to find the relationship between missing values and Sales Price

```python
#if there is a missing value we can drop it but before that try to understand if there is a relationship between these 
#missing values and dependent feature
```

```python
for feature in feature_with_na:
    data=dataset.copy()
    data[feature] = np.where(data[feature].isnull(), 1, 0)  
    
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
    
```

Here With the relation between the missing values and the dependent variable is clearly visible.So We need to replace these nan values with something meaningful which we will do in the Feature Engineering section

From the above dataset some of the features like Id is not required

## Numerical Variables

```python
numerical_features=[feature for feature in dataset.columns if dataset[feature].dtype != 'O']  #O means Object means string value

print("Number of numerical variables:",len(numerical_features))

dataset[numerical_features].head()
```

### Temporal Variables(Eg: Datetime Variables)


```python
year_feature=[feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
dataset[year_feature].head()
```

```python
# let's explore the content of these year variables
for feature in year_feature:
    print(feature, dataset[feature].unique())
```

```python
## Lets analyze the Temporal Datetime Variables
## We will check whether there is a relation between year the house is sold and the sales price
for feature in year_feature:
    dataset.groupby(feature)['SalePrice'].median().plot()
    plt.xlabel(feature)
    plt.ylabel('Median House Price')
    plt.title("House Price vs YearSold")
    plt.show()
    
```

here in year sold the price is decreasing as year is increasing so this cant be true so we will find more info abt YrSold

```python
year_feature
```

```python
## Here we will compare the difference between All years feature with SalePrice

for feature in year_feature:
    if feature != 'YrSold':
        data=dataset.copy()
         ## We will capture the difference between year variable and year the house was sold for
        data[feature]=data['YrSold']-data[feature]
        
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('Median House Price')
        plt.show()
        
```

```python
discrete_feature=[feature for feature in numerical_features if len(dataset[feature].unique())<25 and feature not in year_feature+["Id"]]
print ("Discrete Variables Count: {}".format(len(discrete_feature)))
```

###  Continous Variable

```python
continuous_feature=[feature for feature in numerical_features if feature not in discrete_feature + year_feature+['Id']]
print("Continuous feature Count {}".format(len(continuous_feature)))
```

```python
dataset[continuous_feature].head()
```

```python
## Lets analyse the continuous values by creating histograms to understand the distribution

for feature in continuous_feature:
    data=dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()

```

```python
for feature in continuous_feature:
    data=dataset.copy()  
    if 0 in data[feature].unique():    
        pass
    else:
        data[feature]=np.log(data[feature])
        data["SalePRice"]=np.log(data['SalePrice'])  #coz it is somwhat skewed
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalesPrice')
        plt.title(feature)
        plt.show()
        
```

### Outliers

```python
for feature in continuous_feature:
    data=dataset.copy()  
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature) 
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
